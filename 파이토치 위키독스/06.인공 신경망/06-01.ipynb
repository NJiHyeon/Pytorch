#### 1. 머신러닝 모델의 평가

- 검증용 데이터 : 모델의 성능을 평가하기 위한 용도가 아니라 모델의 성능을 조절하기 위한 용도이다. 더 정확히는 과적합이 되고 있는지 판단하거나 하이퍼파라미터의 조정을 위한 용도

        - ex) 학습률(learning rate), 은닉층의 수, 뉴런의 수, 드롭아웃 비율

#### 2. 분류와 회귀
- 머신러닝의 많은 문제는 분류 또는 회귀 문제에 속한다. 앞서 머신러닝 기법 중 선형회귀와 로지스틱 회귀를 공부했는데 선형 회귀를 통해 회귀 문제에 대해서 학습하고, 로지스틱 회귀를 통해 분류 문제를 학습했다. 
- 분류는 또한 이진분류와 다중 클래스 분류로 나눈다.

##### 1) 이진 분류 문제(Binary Classification)
주어진 입력에 대해서 둘 중 하나의 답을 정하는 문제이다. 
- ex) 시험 성적에 대해서 합격, 불합격인지 판단, 메일로부터 정상 메일, 스팸 메일인지를 판단하는 문제 등

##### 2) 다중 클래스 분류(Multi-class Classification)
주어진 입력에 대해서 세 개 이상의 정해진 선택지 중에서 답을 정하는 문제이다.
- ex) 서점 아르바이트를 하는데 과학, 영어, IT, 학습지, 만화라는 레이블이 각각 붙여져 있는 5개의 책장이 있다고 할 때, 새 책이 입고되면 다섯 개의 책장 중에서 분야에 맞는 적절한 책장에 책을 넣어야 한다. 
    - 이때, 다섯 개의 선택지를 주고 카테고리 또는 범주 또는 클래스라고 한다.
    - 주어진 입력으로부터 정해진 클래스 중 하나로 판단하는 것을 다중 클래스 분류 문제라고 한다. 

##### 3) 회귀 문제(Regression)
분류 문제처럼 0 또는 1이나 과학책장/IT책장 등과 같이 분리된(비연속적인) 답이 결과가 아니라 연속된 값을 결과로 가진다.
- ex) 시험 성적을 예측하는데 5시간 공부하였을 때 80점, 5시간 1분 공부하였을 때 80.5점, 7시간 공부하였을 때는 90점 등이 나오는 것과 같은 문제
- ex) 시계열 데이터를 이용한 주가 예측, 생산량 예측, 지수 예측 등이 있다.

#### 3. 지도학습과 비지도 학습

##### 1) 지도학습 : 레이블이라는 정답과 함께 학습하는 것
- 레이블이라는 말 외에도 y, 실제값 등으로 부르기도 하는데 이 책에서는 이 용어들을 상황에 따라 바꿔서 사용
- 이때, 기계는 예측값과 실제값의 차이인 오차를 줄이는 방식으로 학습을 하게 되는데 예측값은 y^hat과 같이 표현하기도 한다.


##### 2) 비지도학습 : 기본적으로 목적 데이터(또는 레이블)이 없는 학습 방법
- 대표적으로 군집(clustering)이나 차원 축소와 같은 학습 방법들을 비지도 학습이라고 한다. 

##### 3) 강화학습 : 어떤 환경 내에서 정의된 에이전트가 현재의 상태를 인식하여 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 행동 순서를 선택하는 방법

#### 5. 혼돈 행렬(Confusion Matrix)
- Accuracy : 맞춘 문제수를 전체 문제수로 나눈 값
    - 정확도는 맞춘 결과와 틀린 결과에 대한 세부적인 내용을 알려주지는 않는다.
    - 이를 위해서 사용하는 것이 혼돈행렬이다.
- 예를 들어 양성(Positive)와 음성(Negative)을 구분하는 이진 분류가 있다고 하였을 때 혼돈행렬은 다음과 같다.
    - 각 행은 실제값이고, 각 열은 예측값을 나타낸다. 
    - 이를 각각 TP(Ture Positive), TN(True Negative), FP(False Positive), FN(False Negative)
        - True는 정답을 맞춘 경우, Flase는 정답을 맞추지 못한 경우
        - Positive, Negative는 각각 제시했던 정답
        - 즉, TP는 양성이라고 대답하였고 실제로 양성이라서 정답을 맞춘경우, TN은 음성이라고 대답하였고 실제로 음성이라서 정답을 맞춘경우, FP는 양성이라고 대답하였는데 음성이라고 대답하여 틀린 경우, FN은 음성이라고 대답하였는데 실제로 양성이라서 틀린 경우

```python
Image("m3.PNG")
```

##### 1) 정밀도(Precision) : 양성이라고 대답한 전체 케이스에 대한 TP의 비율
- 정밀도 = TP/(TP+FP)

##### 2) 재현률(Recall) : 실제값이 양성인 데이터의 전체 개수에 대한 TP의 비율(양성인 데이터 중에서 얼마나 양성인지를 예측했는지를 나타낸다)
- 재현률 = TP/(TP+FN)

#### 6. 과적합(Overfitting)과 과소적합(Underfitting)

- 과적합 : 훈련 데이터를 과하게 학습한 경우
    - 훈련 데이터는 실제로 존재하는 많은 데이터의 일부에 불과하다.
    - 그런데기계가 훈련 데이터에 대해서만 과하게 학습하면 테스트 데이터나 실제 서비스에서의 데이터에 대해서는 정확도가 좋지 않은 현상이 발생
    - ex) 강아지사진과 고양이사진을 구분하는 기계가 있을 때, 검은색 강아지 사진 훈련 데이터를 과하게 학습하면 기계는 나중에 가서는 흰색 강아지나, 갈색 강아지를 보고도 강아지가 아니라고 판단하여 훈련 데이터에 대해 지나친 일반화를 한 상황이다. 
    - 과적합 상황에서는 훈련 데이터에 대해서는 오차가 낮지만 테스트 데이터에 대해서는 오차가 높아지는 상황이 발생한다.(아래의 그래프 참고)

```python
Image("m4.PNG")
```

- X축의 에포크 : 전체 훈련 데이터에 대한 훈련 횟수를 의미
    - 에포크가 3~4를 넘어가게 되면 과적합 발생
    - 위의 그래프는 테스트 데이터에 대한 오차가 점차 증가하는 양상을 보여준다. 
    - 반대로 말하면 훈련 데이터에 대한 정확도는 높지만 테스트 데이터는 정확도가 낮은 상황이라고 말할 수도 있다. 
    - 즉, 테스트 데이터의 오차가 증가하기 전이나, 정확도가 감소하기 전에 훈련을 멈추는 것이 바람직하다.
- 과소적합
    - 테스트 데이터의 성능이 낮아지기 전에 훈련을 멈추는 것이 바람직하다고 했는데, 테스트 데이터의 성능이 올라갈 여지가 있으에도 훈련을 덜 한 상태를 반대로 과소적합이라고 한다.
    - 과소 적합은 훈련 자체가 부족한 상태이므로 과대 적합과는 달리 훈련 데이터에 대해서도 보통 정확도가 낮다는 특징이 있다.
- 두 가지 현상의 의미 : 머신 러닝에서 학습 또는 훈련이라고 하는 과정을 적합(fitting)이라고 부를 수 있기 때문에 과소/과대 적합이라고 부른다. 즉, 모델이 주어진 데이터에 대해서 적합해져가는 과정이기 때문이다.
- 딥러닝을 할 때는 과적합을 막을 수 있는 드롭아웃, 조기종료와 같은 몇 가지 방법이 존재한다.
