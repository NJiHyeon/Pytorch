### 01. 원-핫 인코딩

#### 1. 원-핫 인코딩이란 ?
- 원-핫 인코딩 : 선택해야 하는 선택지의 개수만큼의 차원을 가지면서, 각 선택지의 인덱스에 해당하는 원소에는 1, 나머지 원소는 0의 값을 가지도록 하는 표현 방법
    - ex) 강아지, 고양이, 냉장고라는 3개의 선택지가 있다고 했을 때
    - 우선 각 선택지에 순차적으로 정수 인덱스를 부여한다. (강아지는 0번 인덱스, 고양이는 1번 인덱스, 냉장고는 2번 인덱스를 부여)
        - 강아지 = [1, 0, 0]
        - 고양이 = [0, 1, 0]
        - 냉장고 = [0, 0, 1]
    - 총 선택지는 3개였으므로 위 벡터들은 전부 3차원의 벡터가 된다. 또한 각 선택지의 벡터들을 보면 해당 선택지의 인덱스에만 1의 값을 가지고, 나머지 원소들은 0의 값을 가진다.
    - 이와 같이 원-핫 인코딩으로 표현된 벡터를 원-핫 벡터라고 한다.

#### 2. 원-핫 벡터의 무작위성
- 꼭 실제값을 원-핫 벡터로 표현해야만 다중 클래스 분류 문제를 풀 수 있는 것은 아니지만, 대부분의 다중 클래스 분류 문제가 각 클래스 간의 관계가 균등하다는 점에서 원-핫 벡터는 이러한 점을 표현할 수 이쓴ㄴ 적절한 표현 방법이다.

- 다수의 클래스를 분류하는 문제에서는 이진 분류처럼 2개의 숫자 레이블이 아니라 클래스의 개수만큼 숫자 레이블이 필요하다.
    - 레이블링 방법은 분류해야 할 클래스 전체에 정수 인코딩을 하는 것을 생각
    - 예를 들어 분류해야 할 레이블이 {red, green, blue}와 같이 3개라면 각각 0, 1, 2로 레이블
    - 그런데 일반적인 다중 클래스 분류 문제에서 레이블링 방법으로는 위와 같은 정수 인코딩이 아니라 원-핫 인코딩을 사용하는 것이 클래스의 성질을 잘 표현했다고 할 수 있다. 
    - 그 이유는 ?

[example]
- Banana, Tomato, Apple 이라는 3개의 클래스가 존재하는 문제가 있을 때, 레이블은 정수 인코딩을 사용하여 각각 1, 2, 3을 부여
    - 손실 함수로 MSE를 사용 
        - 실제값이 Tomato일 때 예측값이 Banana였다면 제곱 오차 = (2-1)^2 = 1
        - 실제값이 Apple일 때 예측값이 Banana였다면 제곱 오차 = (3-1)^2 = 4
        - 즉, Banana와 Tomato 사이의 오차보다 Banana와 Apple의 오차가 더 크다. 이는 컴퓨터에 Banana가 Apple보다는 Tomato에 더 가깝다는 정보를 주는 것과 다름없다. 
        - 순서의 의미를 갖고 있어서 회귀를 통해서 분류 문제를 풀 수 있는 경우에는 도움되지만 여기서는 모든 클래스 간의 관계를 균등하게 분배하는 것이 필요하다. 따라서 원-핫 인코딩을 사용
- 따라서 원-핫 벡터는 각 클래스의 표현 방법이 무작위성을 가진다는 점을 표현할 수 있다. 때로는 단어의 유사성을 구할 수 없다는 단점으로 언급되기도 한다.
